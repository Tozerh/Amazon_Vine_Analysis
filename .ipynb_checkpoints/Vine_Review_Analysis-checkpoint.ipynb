```python
# import filtered table for counts
filter_version = pd.read_csv('filtered.csv')
```

```python
# determines tot. number of paid reviews
paid_tot = filter_version[filter_version["vine"]=="Y"]
paid_tot["vine"].count()
```

```python
# determines tot. number of unpaid reviews
unpaid_tot = filter_version[filter_version["vine"]=="N"]
unpaid_tot["vine"].count()
```

```python
# determines tot. number of paid 5-star reviews
paid_5star = paid_tot[paid_tot['star_rating'] == 5]
paid_5star["vine"].count()
```

```python
# determines tot. number of unpaid 5-star reviews
unpaid_5star = unpaid_tot[unpaid_tot['star_rating'] == 5]
unpaid_5star["vine"].count()
```

```python
# determines percentage of paid 5-star reviews
paid_5star = paid_tot[paid_tot['star_rating'] == 5]
paid_5_percentage = (paid_5star["vine"].count()/paid_tot["vine"].count())
paid_5_percentage
```

```python
# determines percentage of unpaid 5-star reviews
unpaid_5star = unpaid_tot[unpaid_tot['star_rating'] == 5]
unpaid_5_percentage = (unpaid_5star["vine"].count()/unpaid_tot["vine"].count())
unpaid_5_percentage
```
